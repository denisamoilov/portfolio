## –ü–æ—Å—Ç—Ä–æ–∏–º –º–æ–¥–µ–ª—å –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏—è –æ—Ç—Ç–æ–∫–∞ –∫–ª–∏–µ–Ω—Ç–æ–≤

**–ü–æ—Å—Ç—Ä–æ–∏–º –º–æ–¥–µ–ª—å –±–∏–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –∫–ª–∏–µ–Ω—Ç–æ–≤, –≥–¥–µ —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ ‚Äî —Ñ–∞–∫—Ç –æ—Ç—Ç–æ–∫–∞ –∫–ª–∏–µ–Ω—Ç–∞ –≤ —Å–ª–µ–¥—É—é—â–µ–º –º–µ—Å—è—Ü–µ**

### –†–∞–∑–æ–±—å–µ–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –≤—ã–±–æ—Ä–∫—É —Ñ—É–Ω–∫—Ü–∏–µ–π train_test_split()

<div class="alert alert-warning">

**–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞**

–ú–æ–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –ø–∞—Ä–∞–º–µ—Ç—Ä stratify = y –≤ —Ñ—É–Ω–∫—Ü–∏–∏ train_test_split( ), –≥–¥–µ y ‚Äì —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∞—è —Ü–µ–ª–µ–≤–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è. –î–∞–Ω–Ω—ã–µ —Ä–∞–∑–¥–µ–ª—è—Ç—Å—è –Ω–∞ –¥–≤–∞ –∫–ª–∞—Å—Ç–µ—Ä–∞ (–Ω–∞–±–ª—é–¥–µ–Ω–∏—è –∫–ª–∞—Å—Å–∞ 0 –∏ –Ω–∞–±–ª—é–¥–µ–Ω–∏—è –∫–ª–∞—Å—Å–∞ 1), –∏ –≤ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –≤—ã–±–æ—Ä–∫—É –ø–æ–ø–∞–¥—É—Ç –ø–æ 20% –Ω–∞–±–ª—é–¥–µ–Ω–∏–π –∏–∑ –∫–∞–∂–¥–æ–≥–æ –∫–ª–∞—Å—Ç–µ—Ä–∞:


–¢–æ –µ—Å—Ç—å, –µ—Å–ª–∏ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤ –≤ –æ—Ä–∏–≥–∏–Ω–∞–ª—å–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ –±—ã–ª–æ 2:5, —Ç–æ –≤ –æ–±—É—á–∞—é—â–µ–π –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∞—Ö –æ–Ω–æ —Å–æ—Ö—Ä–∞–Ω–∏—Ç—Å—è —Ç–∞–∫–∏–º –∂–µ. –¢–∞–∫–æ–π –ø–æ–¥—Ö–æ–¥ –≤ —á–µ–º-—Ç–æ —Ä–µ—à–∞–µ—Ç –ø—Ä–æ–±–ª–µ–º—É –Ω–µ—Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –∫–ª–∞—Å—Å–æ–≤.

–§—É–Ω–∫—Ü–∏—è train_test_split( ) —Å—Ç—Ä–æ–µ–Ω–∞ —Ç–∞–∫, —á—Ç–æ –º—ã –Ω–µ —Å—Ç–∞–ª–∫–∏–≤–∞–µ–º—Å—è —Å –ø–æ–¥–æ–±–Ω–æ–π –ø—Ä–æ–±–ª–µ–º–æ–π. –ù–æ —è –≤—Å–µ –∂–µ –ø–∏—à—É –ø—Ä–æ —ç—Ç–æ—Ç –ø–∞—Ä–∞–º–µ—Ç—Ä, —á—Ç–æ–±—ã –æ–±—Ä–∞—Ç–∏—Ç—å –≤–Ω–∏–º–∞–Ω–∏–µ –Ω–∞ –ø—Ä–æ–±–ª–µ–º—É –¥–∏—Å–±–∞–ª–∞–Ω—Å–∞ :)</div>

```python
X = data.drop(['churn'], axis = 1)
y = data['churn']
#—Ä–∞–∑–æ–±—å–µ–º –≤—ã–±–æ—Ä–∫–∏ –≤ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–∏ 80 –Ω–∞ 20 %%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### –û–±—É—á–∏–º –º–æ–¥–µ–ª—å –Ω–∞ train-–≤—ã–±–æ—Ä–∫–µ –¥–≤—É–º—è —Å–ø–æ—Å–æ–±–∞–º–∏: –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–µ–π –∏ —Å–ª—É—á–∞–π–Ω—ã–º –ª–µ—Å–æ–º

```python
#–Ω–æ—Ä–º–∞–ª–∏–∑—É–µ–º –≤—ã–±–æ—Ä–∫–∏
scaler = StandardScaler()
X_train_st = scaler.fit_transform(X_train)
X_test_st = scaler.transform(X_test)

#–æ–±—É—á–∏–º –º–æ–¥–µ–ª—å –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–µ–π
lr_model = LogisticRegression(random_state=0)
lr_model.fit(X_train_st, y_train)
lr_predictions = lr_model.predict(X_test_st)
lr_probabilities = lr_model.predict_proba(X_test_st)[:,1]

#–æ–±—É—á–∏–º –º–æ–¥–µ–ª—å —Å–ª—É—á–∞–π–Ω—ã–º –ª–µ—Å–æ–º
rf_model = RandomForestClassifier(n_estimators = 100, random_state = 0) 
rf_model.fit(X_train_st, y_train)
rf_predictions = rf_model.predict(X_test_st)
rf_probabilities = rf_model.predict_proba(X_test_st)[:,1]
```

### –û—Ü–µ–Ω–∏–º –º–µ—Ç—Ä–∏–∫–∏ accuracy, precision –∏ recall –¥–ª—è –æ–±–µ–∏—Ö –º–æ–¥–µ–ª–µ–π –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ. –°—Ä–∞–≤–Ω–∏–º –ø–æ –Ω–∏–º –º–æ–¥–µ–ª–∏. –û–ø—Ä–µ–¥–µ–ª–∏–º, –∫–∞–∫–∞—è –º–æ–¥–µ–ª—å –ø–æ–∫–∞–∑–∞–ª–∞ —Å–µ–±—è –ª—É—á—à–µ –Ω–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –º–µ—Ç—Ä–∏–∫.

```python
#–≤—ã–≤–µ–¥–µ–º –º–µ—Ç—Ä–∏–∫–∏ –¥–ª—è –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏, –ø—Ä–æ—Ç–µ—Å—Ç–∏—Ä—É–µ–º –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
print('–ú–µ—Ç—Ä–∏–∫–∏ –¥–ª—è –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏:')
print('accuracy_score: {}\nprecision_score: {}\nrecall_score:{}'.format(
    accuracy_score(y_test, lr_predictions),
    precision_score(y_test, lr_predictions),
    recall_score(y_test, lr_predictions)
))

#–≤—ã–≤–µ–¥–µ–º –º–µ—Ç—Ä–∏–∫–∏ –¥–ª—è —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞, –ø—Ä–æ—Ç–µ—Å—Ç–∏—Ä—É–µ–º –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
print('\n–ú–µ—Ç—Ä–∏–∫–∏ –¥–ª—è —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞:')
print('accuracy_score: {}\nprecision_score: {}\nrecall_score:{}'.format(
    accuracy_score(y_test, rf_predictions),
    precision_score(y_test, rf_predictions),
    recall_score(y_test, rf_predictions)
))
```

```python
X_st = scaler.fit_transform(X)
```

### –ü–æ—Å—Ç—Ä–æ–∏–º –º–∞—Ç—Ä–∏—Ü—É —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–π —Ñ—É–Ω–∫—Ü–∏–µ–π linkage() –Ω–∞ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–æ–≤–∞–Ω–Ω–æ–π –º–∞—Ç—Ä–∏—Ü–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏ –Ω–∞—Ä–∏—Å—É–µ–º –¥–µ–Ω–¥—Ä–æ–≥—Ä–∞–º–º—É

### –û–±—É—á–∏–º –º–æ–¥–µ–ª—å –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏ –Ω–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –∞–ª–≥–æ—Ä–∏—Ç–º–∞ K-Means –∏ —Å–ø—Ä–æ–≥–Ω–æ–∑–∏—Ä—É–π—Ç–µ –∫–ª–∞—Å—Ç–µ—Ä—ã –∫–ª–∏–µ–Ω—Ç–æ–≤

```python
print('–î–æ–ª—è –æ—Ç—Ç–æ–∫–∞ –ø–æ –∫–ª–∞—Å—Ç–µ—Ä–∞–º')
display((data_clusters['churn'] *100).round(1).astype(str) + '%')
```
