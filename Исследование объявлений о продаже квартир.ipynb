### Шаг 2. Предобработка данных

```python
#произведем замену пропусков и типов данных в столбцах:
#предлагаю заменить пропущенные значения с столбцах зменить на медианные значения, что позволит нам сохранить ценные данныы,
#а также обойтись без замены пропущенных значений на 0, что может повлиять на результаты анализа.
#На 0 произведем замену только в столбцах с информацией о наличии в радиусе 3 км парков и водоемов, потому что их может не быть.
#Оставим пропуски в данных о расстоянии до аэропорта, центра города и т.д., так как эти данные формируются автоматически,
#они могли не попасть из-за указания адреса в объявлении, либо неверном указании адреса. Замена пропусков на значения может повлиять
#на выборку при при анализе данных, поэтому оставим их как есть. 
#Тип данных меняем при помощи метода astype()
data['last_price'] = data['last_price'].astype('float')
data['total_area'] = data['total_area'].astype('float')
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median()).astype('float')
data['airports_nearest'] = data['airports_nearest'].astype('float')
data['cityCenters_nearest'] = data['cityCenters_nearest'].astype('float')
data['parks_nearest'] = data['parks_nearest'].astype('float')
data['ponds_nearest'] = data['last_price'].astype('float')

data['balcony'] = data['balcony'].fillna(0).astype('float').astype('int')
data['parks_around3000'] = data['parks_around3000'].fillna(0).astype('float').astype('int')
data['ponds_around3000'] = data['ponds_around3000'].fillna(0).astype('float').astype('int')
data['floors_total'] = data['floors_total'].fillna(data['floors_total'].median()).astype('float').astype('int')
  
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y.%m.%d')

#ИСПРАВЛЕНИЕ ЗАМЕНЫ ПРОПУСКОВ!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
#(print(data['living_area'].corr(data['total_area']),
#      data['living_area'].corr(data['rooms']),
#      data['living_area'].corr(data['last_price']))
#)

#обнаружена высокая корреляция жилой площади с общей площадью, она составляет 0,87. 
#можно посчитать долю жилой площади от общей и применить этот коэффициент к общей площади, чтобы найти
#пропущенные значения в жилой. Добавим соответсвующий столбец в таблицу
data['living_area'] = data['living_area'].astype('float')
data['living_area_part'] = data['living_area']/data['total_area']
#display(data)
#в новый столбец перехали пропуски из столца с информацией о жилой площади, здесь можно заменить пропуски на среднее значение
data['living_area_part'] = data['living_area_part'].fillna(data['living_area_part'].mean())
#print(data['living_area_part'].isnull().sum())
#применим полученные коэффициенты к столбцу с информацией об общей площади и сообщим это в качестве аргемента методу замены пропусков
data['living_area'] = data['living_area'].fillna(data['living_area_part'] * data['total_area'])
#проверим пропуски
#print(data['living_area'].isnull().sum())

#проделаем те же операции с площадью кухни
data['kitchen_area'] = data['kitchen_area'].astype('float')
#(print(data['kitchen_area'].corr(data['total_area']),
#      data['kitchen_area'].corr(data['last_price']),
#      data['kitchen_area'].corr(data['last_price']))
#)
#также обнаружена корреляция между площадью кухни и общей площадью, не такая же высокая, но также существенная 0,61
data['kitchen_area_part'] = data['kitchen_area']/data['total_area']
#display(data)
#в новый столбец перехали пропуски из столца с информацией о жилой площади, здесь можно заменить пропуски на среднее значение
data['kitchen_area_part'] = data['kitchen_area_part'].fillna(data['kitchen_area_part'].mean())
#print(data['living_area_part'].isnull().sum())
#применим полученные коэффициенты к столбцу с информацией об общей площади и сообщим это в качестве аргемента методу замены пропусков
data['kitchen_area'] = data['kitchen_area'].fillna(data['kitchen_area_part'] * data['total_area'])
#проверим пропуски
#print(data['kitchen_area'].isnull().sum())

#замена пропусков и типа данных в days_exposition
#проверим, когда возникли пропуски, для этого добавим новый столбец и посчитаем количество пропусков
#затем построим график
#data['days_exposition_missing'] = data['days_exposition'].isna().astype('int')
#data.groupby('first_day_exposition').agg({'days_exposition_missing' : 'sum'}).plot()
#plt.show()
#на графике видно, что количество пропусков резко увеличивается к концу периода.
#это, скорее всего, означает, что объявления были активны на момент выгрузки данных
#для замены пропусков можно принять что дата снятия продажи(снятия с публикации) равна последнему дню в наборе данных
max_dt = data['first_day_exposition'].max()
data.loc[data['days_exposition'].isna(), 'days_exposition'] = (max_dt - data.loc[data['days_exposition'].isna(), 'first_day_exposition']).dt.days 
data['days_exposition'] = data['days_exposition'].astype('float').astype('int')

#параметр апартаментов указывается сами продавцом. Так как такой тип жилья появился недавно, в старых объявления его могло не быть
#поэтому заполни все пропуски значением False
data['is_apartment'] = data['is_apartment'].fillna('False')

#разберемся с большими значениями
#data['rooms'].unique()
data_rooms_WOW = data.query('rooms >= 10')
display(data_rooms_WOW)
data['total_area'].corr(data['rooms'])
#большое количество комнат соответсвует большим по площадям помещениям, поэтому скорее всего не является ошибкой
#дополнительно проверим корреляцию этих двух параметров, она составляет 0,75, что подтверждает мои предположения

#проверим, что там с потолками
data['ceiling_height'].unique()
data_ceiling_height_WOW = data.query('ceiling_height >= 10')
data_ceiling_height_WOW
data['ceiling_height'].describe()
#есть два десятка очень высоких значений по высоте потолков, среди них есть апартаменты, т.е. возможно, это жилье
#типа лофт в реновируемых промышленных зонах. Единственное значение, которое кажется странным это 100 метров.
#Однако, при проверке методом describe этот выброс не сильно влияет на среднюю и медиану. 
data.info()
#data.head()
```
