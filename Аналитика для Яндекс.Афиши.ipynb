<div class="alert alert-danger">

**–í—ã–≤–æ–¥**

–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –ø–æ—Å–µ—Ç–∏–≤—à–∏—Ö —Å–∞–π—Ç –∑–∞ –º–µ—Å—è—Ü (mau): 23228 —á–µ–ª–æ–≤–µ–∫.–ù–∞–∏–±–æ–ª–µ–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã–º –æ–∫–∞–∑–∞–ª—Å—è –Ω–æ—è–±—Ä—å 2017 –≥–æ–¥–∞. –¢–∞–∫–∂–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã –∑–∏–º–Ω–∏–µ –º–µ—Å—è—Ü—ã: –¥–µ–∫–∞–±—Ä—å 2017 –∏ —è–Ω–≤–∞—Ä—å 2018 –≥–≥, –¥–∞–ª–µ–µ –Ω–∞–±–ª—é–¥–∞–µ—Ç—Å—è –ø–æ—Å—Ç–µ–ø–µ–Ω–Ω–æ–µ —Å–Ω–∏–∂–µ–Ω–∏–µ –≤–∏–∑–∏—Ç–æ–≤, –º–∏–Ω–∏–º—É–º –ø—Ä–∏—Ö–æ–¥–∏—Ç—Å—è –Ω–∞ –ª–µ—Ç–Ω–∏–µ –º–µ—Å—è—Ü—ã.


<div class="alert alert-danger">
<s><b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞:</b> –ê–Ω–∞–ª–æ–≥–∏—á–Ω–∞—è –ø—Ä–æ–±–ª–µ–º–∞. –ò—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ —à–ª–æ  —Å –∏—é–Ω—è –ø–æ –º–∞–π, –∞ –Ω–µ —Å —è–Ω–≤–∞—Ä—è –ø–æ –¥–µ–∫–∞–±—Ä—å.
</div>

<div class="alert alert-info"> <b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Å—Ç—É–¥–µ–Ω—Ç–∞:</b> –ò—Å–ø—Ä–∞–≤–∏–ª, —Å–ø–∞—Å–∏–±–æ</div>

```python
#–æ–ø—Ä–µ–¥–µ–ª–∏–º, —Å–∫–æ–ª—å–∫–æ –≤ —Å—Ä–µ–¥–Ω–µ–º —Ä–∞–∑ –≤ –¥–µ–Ω—å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ –∑–∞—Ö–æ–¥—è—Ç –Ω–∞ —Å–∞–π—Ç
#–ø–æ—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –¥–µ–Ω—å
users_for_day_uniq = visits.groupby('date').agg({'uid': 'nunique'}).reset_index()

#–ø–æ—Å—á–∏—Ç–∞–µ–º –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –¥–µ–Ω—å
users_for_day_count = visits.groupby('date').agg({'uid': 'count'}).reset_index()

#–æ–±—ä–µ–¥–∏–Ω–∏–º –¥–∞–Ω–Ω—ã–µ –∏ –ø–æ—Å—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–∏–∑–∏—Ç–æ–≤ –¥–µ–Ω—å
users_for_day = users_for_day_uniq.merge(users_for_day_count, on = 'date')
users_for_day.columns = ['date', 'uniq_uid', 'count_visits']
users_for_day['mean_visits'] = users_for_day['count_visits'] / users_for_day['uniq_uid']

display(users_for_day.head(15))
print("–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–∏–∑–∏—Ç–æ–≤ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –Ω–∞ —Å–∞–π—Ç –≤ –¥–µ–Ω—å", users_for_day['mean_visits'].mean())

#–ø–æ—Å—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–∏–∑–∏—Ç–æ–≤ –≤ –¥–µ–Ω—å –≤ —Ä–∞–∑—Ä–µ–∑–µ –º–µ—Å—è—Ü–µ–≤, —Ç–∞–∫ –∫–∞–∫ –æ–Ω–∏ –Ω–µ—Ä–∞–≤–Ω–æ–≤–µ—Å–Ω—ã
users_for_day['visit_month'] = users_for_day['date'].astype('datetime64[M]')
users_for_day_month = users_for_day.groupby('visit_month').agg({'uniq_uid': 'sum', 'count_visits': 'sum'}).reset_index()
users_for_day_month['mean_visits'] = users_for_day_month['count_visits'] / users_for_day_month['uniq_uid']
display(users_for_day_month.head(15))

#–ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫
users_for_day_month['visit_month'] = users_for_day_month['visit_month'].astype('str')
sns.set()
plt.figure(figsize = (10,15))
ax = sns.barplot(data = users_for_day_month, x = 'mean_visits', y = 'visit_month', color = 'salmon', orient = "h")
plt.xlabel("–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–∏–∑–∏—Ç–æ–≤ –Ω–∞ —Å–∞–π—Ç", fontsize = 15)
plt.ylabel("–î–∞—Ç–∞", fontsize = 15)
plt.title("–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–∏–∑–∏—Ç–æ–≤ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –Ω–∞ —Å–∞–π—Ç –≤ –¥–µ–Ω—å", fontsize = 15)

for i in ax.patches:
    ax.text(i.get_width()+.1, i.get_y()+.31, \
            str(round((i.get_width()), 2)), fontsize=15, color='dimgrey')
ax.set_xticks([0, 2])
```

**–í—ã–≤–æ–¥**

–ì—Ä–∞—Ñ–∏–∫ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∏–º–µ–µ—Ç —Å–º–µ—â–µ–Ω–∏–µ –≤–ª–µ–≤–æ –∏ –¥–ª–∏–Ω–Ω—ã–π —Ö–≤–æ—Å—Ç. –ü—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å —Ç–∏–ø–∏—á–Ω–æ–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–æ–π —Å–µ—Å—Å–∏–∏ –∑–∞ –≤–µ—Å—å –ø–µ—Ä–∏–æ–¥ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç 1 –º–∏–Ω—É—Ç—É. –ë–æ–ª—å—à–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –ø—Ä–æ–≤–æ–¥—è—â–∏—Ö –Ω–∞ —Å–∞–π—Ç–µ –Ω–µ –±–æ–ª–µ–µ 1 –º–∏–Ω—É—Ç—ã –º–æ–∂–µ—Ç —Å–≤–∏–¥–µ—Ç–µ–ª—å—Å—Ç–≤–æ–≤–∞—Ç—å –æ –Ω–µ–ø–æ–ª–∞–¥–∫–∞—Ö –≤ —Å–µ—Ä–≤–∏—Å–µ, –æ—Å–ª–æ–∂–Ω–µ–Ω–Ω–æ–π –Ω–∞–≤–∏–≥–∞—Ü–∏–∏ (–ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å –Ω–µ –≤–∏–¥–∏—Ç —Å—Ä–∞–∑—É —Ç–æ, —á—Ç–æ —Ö–æ—á–µ—Ç) –∏ –∫–∞–∫ —Å–ª–µ–¥—Å—Ç–≤–∏–µ —Å–Ω–∏–∂–µ–Ω–∏–∏ –∏–Ω—Ç–µ—Ä–µ—Å–∞ –∫ —Å–µ—Ä–≤–∏—Å—É. 

<div class="alert alert-info"> <b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Å—Ç—É–¥–µ–Ω—Ç–∞:</b> –ò—Å–ø—Ä–∞–≤–∏–ª </div>

```python
#–†–∞—Å—Å—á–∏—Ç–∞–µ–º Retention Rate, –ø—Ä–∏–º–µ–Ω—è—è –∫–æ–≥–æ—Ä—Ç–Ω—ã–π –∞–Ω–∞–ª–∏–∑.
#–≤—ã–¥–µ–ª–∏–º –¥–∞—Ç—É —Å–∞–º–æ–π —Ä–∞–Ω–Ω–µ–π –ø–æ–∫—É–ø–∫–∏
first_visit_day = visits.groupby('uid')['start ts'].min()
first_visit_day.name = 'first_visit_date'
visits_1 = visits.merge(first_visit_day, on = 'uid')
display(visits_1.head())
```

<div class="alert alert-warning">
<s><b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞: </b> –°–æ–≤–µ—Ç: —Å –æ–±—ä–µ–∫—Ç–∞–º–∏ panads —É–¥–æ–±–Ω–µ–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å display() –≤–º–µ—Å—Ç–æ print(). –≠—Ç–æ –¥–∞—Å—Ç —Ç–µ–±–µ –∫—Ä–∞—Å–∏–≤–æ –æ—Ç—Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–Ω—É—é —Ç–∞–±–ª–∏—Ü—É.
</div>

<div class="alert alert-info"> <b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Å—Ç—É–¥–µ–Ω—Ç–∞:</b> –ò—Å–ø—Ä–∞–≤–∏–ª </div>

```python
#–≤—ã–¥–µ–ª–∏–º –º–µ—Å—è—Ü –ø–æ–∫—É–ø–∫–∏ –∏ –º–µ—Å—è—Ü –ø–µ—Ä–≤–æ–π –ø–æ–∫—É–ø–∫–∏
visits_1['first_visit_month'] = visits_1['first_visit_date'].astype('datetime64[M]')
visits_1['visit_month'] = visits_1['start ts'].astype('datetime64[M]')

#—Ä–∞—Å—Å—á–∏—Ç–∞–µ–º lifetime
visits_1['cohort_lifetime'] = ((visits_1['visit_month'] - visits_1['first_visit_month']) / np.timedelta64(1, 'M')).round()

#—Å–æ–∑–¥–∞–¥–∏–º —Ç–∞–±–ª–∏—Ü—É —Å –∫–æ–≥–æ—Ä—Ç–∞–º–∏
cohort = visits_1.groupby(['first_visit_month', 'cohort_lifetime']).agg({'uid': 'nunique'}).reset_index()
#print(cohort)

initial_users = cohort[cohort['cohort_lifetime'] == 0][['first_visit_month', 'uid']]
#print(initial_users) 
initial_users = initial_users.rename(columns={'uid':'cohort_users'})

cohort = cohort.merge(initial_users, on='first_visit_month') 
display(cohort)

#—Ä–∞—Å—Å—á–∏—Ç–∞–µ–º retention
cohort['retention'] = cohort['uid']/cohort['cohort_users']

#–ø–æ—Å—Ç—Ä–æ–∏–º —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É –∏ –ø–æ—Å—á–∏—Ç–∞–µ–º Retention
retention_pivot = cohort.pivot_table(
    index='first_visit_month',
    columns='cohort_lifetime',
    values='retention',
    aggfunc='sum',
)

#–ø–æ—Å—Ç—Ä–æ–∏–º —Ç–µ–ø–ª–æ–≤—É—é –∫–∞—Ä—Ç—É
sns.set(style='white')
plt.figure(figsize=(20, 15))
plt.title('Cohorts: User Retention', fontsize = (30))
sns.heatmap(retention_pivot, annot=True, fmt='.1%', linewidths=1, linecolor='gray')
plt.show()
```

    <li>–î–∏–∞–≥—Ä–∞–º–º—É –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å —è—Ä—á–µ –Ω–∞—Å—Ç—Ä–æ–∏–≤ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã <code>vminv</code>/<code>vmax</code></li>
    <li>–¢–µ–∫—Å—Ç –Ω–∞–¥ –≥—Ä–∞—Ñ–∏–∫–æ–º –º–æ–∂–Ω–æ —Å–∫—Ä—ã—Ç—å —Å –ø–æ–º–æ—â—å—é –¥–æ–±–∞–≤–ª–µ–Ω–∏—è –∫–æ–º–∞–Ω–¥—ã <code>plt.show()</code></li>

```python
#–ù–∞–π–¥–µ–º —Å—Ä–µ–¥–Ω–∏–π Retention Rate –Ω–∞ –≤—Ç–æ—Ä–æ–π –º–µ—Å—è—Ü –∂–∏–∑–Ω–∏ –∫–æ–≥–æ—Ä—Ç
rr2 = retention_pivot[2].mean()
print('—Å—Ä–µ–¥–Ω–∏–π Retention Rate –Ω–∞ –≤—Ç–æ—Ä–æ–π –º–µ—Å—è—Ü ¬´–∂–∏–∑–Ω–∏¬ª –∫–æ–≥–æ—Ä—Ç: {:.1%}'.format(rr2) )
```

```python
#–ò—Å—Å–ª–µ–¥—É–µ–º, —Å–∫–æ–ª—å–∫–æ –≤ —Å—Ä–µ–¥–Ω–µ–º –≤—Ä–µ–º–µ–Ω–∏ –ø—Ä–æ—Ö–æ–¥–∏—Ç —Å –º–æ–º–µ–Ω—Ç–∞ –ø–µ—Ä–≤–æ–≥–æ –ø–æ—Å–µ—â–µ–Ω–∏—è —Å–∞–π—Ç–∞ –¥–æ —Å–æ–≤–µ—Ä—à–µ–Ω–∏—è –ø–æ–∫—É–ø–∫–∏
#–æ–±—ä–µ–¥–∏–Ω–∏–º –¥–∞–Ω–Ω—ã–µ –ø–æ –≤–∏–∑–∏—Ç–∞–º –∏ –ø–æ–∫—É–ø–∫–∞–º
all_data = orders.merge(visits_1, on = 'uid')

#–Ω–∞–π–¥–µ–º –¥–∞—Ç—É –ø–µ—Ä–≤–æ–≥–æ –≤–∏–∑–∏—Ç–∞ –∏ –ø–µ—Ä–≤–æ–π –ø–æ–∫—É–ø–∫–∏
time_to_orders = all_data.groupby('uid').agg({'start ts':'min', 'buy ts': 'min'}).reset_index()

#–Ω–∞–π–¥–µ–º –ø—Ä–æ–º–µ–∂—É—Ç–æ–∫ –º–µ–∂–¥—É –≤—Ä–µ–º–µ–Ω–µ–º –ø–µ—Ä–≤–æ–≥–æ –≤–∏–∑–∏—Ç–∞ –∏ –≤—Ä–µ–º–µ–Ω–µ–º –ø–µ—Ä–≤–æ–π –ø–æ–∫—É–ø–∫–∏
time_to_orders['time_to_order'] = time_to_orders['buy ts'] - time_to_orders['start ts']

#–ø–µ—Ä–µ–≤–µ–¥–µ–º –≤—Ä–µ–º—è –≤ –º–∏–Ω—É—Ç—ã
time_to_orders['time_to_order_sec'] = time_to_orders['time_to_order'].dt.seconds
time_to_orders['time_to_order_min'] = time_to_orders['time_to_order_sec'] / 60
display(time_to_orders.head(15))

#–ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫
plt.figure(figsize = (12,9))
sns.distplot(time_to_orders['time_to_order_min'])
plt.xlabel("–ü—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å —Å–µ—Å—Å–∏–∏, –º–∏–Ω", fontsize = 15)
plt.ylabel("–ü–ª–æ—Ç–Ω–æ—Å—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è", fontsize = 15)
plt.title("–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –ø–µ—Ä–∏–æ–¥–∞ –º–µ–∂–¥—É –ø–µ—Ä–≤—ã–º–∏ –≤–∏–∑–∏—Ç–æ–º –∏ –ø–æ–∫—É–ø–∫–æ–π", fontsize = 15)

#–≤—ã–≤–µ–¥–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ —Å–µ—Å—Å–∏–∏
print("–ú–æ–¥–∞ –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –≤—Ä–µ–º–µ–Ω–∏ –º–µ–∂–¥—É –ø–µ—Ä–≤—ã–º –≤–∏–∑–∏—Ç–æ–º –∏ –ø–µ—Ä–≤–æ–π –ø–æ–∫—É–ø–∫–æ–π", time_to_orders['time_to_order_min'].mode())
print("–°—Ä–µ–¥–Ω—è—è –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –≤—Ä–µ–º–µ–Ω–∏ –º–µ–∂–¥—É –ø–µ—Ä–≤—ã–º –≤–∏–∑–∏—Ç–æ–º –∏ –ø–µ—Ä–≤–æ–π –ø–æ–∫—É–ø–∫–æ–π", time_to_orders['time_to_order_min'].mean())
print("–ú–µ–¥–∏–∞–Ω–Ω–∞—è –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –≤—Ä–µ–º–µ–Ω–∏ –º–µ–∂–¥—É –ø–µ—Ä–≤—ã–º –≤–∏–∑–∏—Ç–æ–º –∏ –ø–µ—Ä–≤–æ–π –ø–æ–∫—É–ø–∫–æ–π", time_to_orders['time_to_order_min'].median())

```

```python
#–†–∞—Å—Å—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–∫—É–ø–æ–∫ –Ω–∞ –æ–¥–Ω–æ–≥–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª—ã –∑–∞ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—ã–π –ø–µ—Ä–∏–æ–¥
#–Ω–∞–π–¥–µ–º –¥–∞—Ç—É –ø–µ—Ä–≤–æ–π –ø–æ–∫—É–ø–∫–∏
first_order_day = orders.groupby('uid')['buy ts'].min()
first_order_day.name = 'first_order_date'

#–¥–æ–±–∞–≤–∏–º –¥–∞—Ç—É –ø–µ—Ä–≤–æ–π –ø–æ–∫—É–ø–∫–∏ –≤ —Ç–∞–±–ª–∏—Ü—É —Å –ø–æ–∫—É–ø–∫–∞–º–∏
orders_1 = orders.merge(first_order_day, on = 'uid')
orders_1['first_order_month'] = orders_1['first_order_date'].astype('datetime64[M]')

#—Å–æ–∑–¥–∞–¥–∏–º —Å—Ç–æ–ª–±–µ—Ü —Å –º–µ—Å—è—Ü–µ–º –ø–æ–∫—É–ø–∫–∏
orders_1['order_month'] = orders_1['buy ts'].astype('datetime64[M]')

#–ø–æ—Å—á–∏—Ç–∞–µ–º –¥–ª—è –∫–∞–∂–¥–æ–π –∫–æ–≥–æ—Ä—Ç—ã –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–∫—É–ø–æ–∫
cohort_orders = orders_1.groupby(['first_order_month', 'order_month']).agg({'revenue': 'count'}).reset_index()
#print(cohort_orders.head())

#—É–∑–Ω–∞–µ–º —Ä–∞–∑–º–µ—Ä –∫–æ–≥–æ—Ä—Ç—ã –≤ –∫–∞–∂–¥—ã–π –º–µ—Å—è—Ü
cohort_sizes = orders_1.groupby(['first_order_month', 'order_month']).agg({'uid': 'nunique'}).reset_index()
cohort_sizes.columns = ['first_order_month', 'order_month', 'n_buyers']
#print(cohort_sizes.head())

#—Å–æ–∑–¥–∞–¥–∏–º –∏—Ç–æ–≥–æ–≤—É—é —Ç–∞–±–ª–∏—Ü—É
report_orders_count = pd.merge(cohort_sizes, cohort_orders, on = ['first_order_month', 'order_month'])
report_orders_count.columns = ['first_order_month', 'order_month', 'n_buyers', 'count_buy']

#—Ä–∞—Å—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–∫—É–ø–æ–∫ –Ω–∞ –æ–¥–Ω–æ–≥–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª—è –≤ –∫–∞–∂–¥–æ–π –∫–æ–≥–æ—Ä—Ç–µ –∑–∞ –∫–∞–∂–¥—ã–π –º–µ—Å—è—Ü
report_orders_count['buy_per_user'] = report_orders_count['count_buy'] / report_orders_count['n_buyers']

print(report_orders_count.head())

#–ø–æ—Å—Ç—Ä–æ–∏–º —Ç–µ–ø–ª–æ–≤—É—é –∫–∞—Ä—Ç—É
count_buy_pivot = report_orders_count.pivot_table(
    index='first_order_month',
    columns='order_month',
    values='buy_per_user',
    aggfunc='mean',
)

sns.set(style='white')
plt.figure(figsize=(20, 15))
plt.title('–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–∫—É–ø–æ–∫ –Ω–∞ –æ–¥–Ω–æ–≥–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è', fontsize = (30))
sns.heatmap(count_buy_pivot, annot=True, linewidths=1, linecolor='gray')

#—Å–æ–∑–¥–∞–¥–∏–º —Å—Ä–µ–∑ –∑–∞ 6 –º–µ—Å—è—Ü–µ–≤ —Å —è–Ω–≤–∞—Ä—è –ø–æ –∏—é–Ω—å 2018 –≥–æ–¥–∞:
report_orders_count_6month = report_orders_count[report_orders_count['order_month'] >= '2018-01-01']
count_buy_pivot_6month = report_orders_count_6month.pivot_table(
    index='order_month',
    values='buy_per_user',
    aggfunc='mean',
)

display(count_buy_pivot_6month)
print('–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–∫—É–ø–æ–∫ –Ω–∞ 1 –ø–æ–∫—É–ø–∞—Ç–µ–ª—è –∑–∞ –ø–µ—Ä–∏–æ–¥ —Å —è–Ω–≤–∞—Ä—è –ø–æ –∏—é–Ω—å 2018 –≥–æ–¥–∞ –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ:', count_buy_pivot_6month['buy_per_user'].mean())
```

–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–∫—É–ø–æ–∫ –Ω–∞ 1 –ø–æ–∫—É–ø–∞—Ç–µ–ª—è –∑–∞ –ø–µ—Ä–∏–æ–¥ —Å —è–Ω–≤–∞—Ä—è –ø–æ –∏—é–Ω—å 2018 –≥–æ–¥–∞ –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç 1.65 –ø–æ–∫—É–ø–∫–∏. –ú–æ–∂–Ω–æ –æ–∫—Ä—É–≥–ª–∏—Ç—å –¥–æ 2 –ø–æ–∫—É–ø–æ–∫. –ü–µ—Ä–≤–∞—è –∫–æ–≥–æ—Ä—Ç–∞ (–∏—é–Ω—è 2017 –≥) –Ω–∞–∏–±–æ–ª–µ–µ –∞–∫—Ç–∏–≤–Ω–∞—è, —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–∫—É–ø–æ–∫ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç 3-4 –ø–æ–∫—É–ø–æ–∫. –ò–Ω—Ç–µ—Ä–µ—Å –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—é—Ç –∫–æ–≥–æ—Ä—Ç—ã –º–∞—Ä—Ç–∞ –∏ –∞–ø—Ä–µ–ª—è 2018 –≥–æ–¥–∞, —Ç–∞–∫ –∫–∞–∫ –æ–Ω–∏ –∞–∫—Ç–∏–≤–Ω—ã –≤ –Ω–µ–ø–æ–ø—É–ª—è—Ä–Ω—ã–π –∏—é–Ω—å 2018 –≥–æ–¥–∞, —Å—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–∫—É–ø–æ–∫ 3. –ò—Å—Ö–æ—è –∏–∑ –ø–æ–≤–µ–¥–µ–Ω–∏—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –∫–æ–≥–æ—Ä—Ç–∞—Ö. –º–æ–∂–Ω–æ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–∏—Ç—å, —á—Ç–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ –∑–∞—Ä–∞–Ω–µ–µ (–∑–∞ 2-3 –º–µ—Å—è—Ü–∞) –ø–ª–∞–Ω–∏—Ä—É—é—Ç –ø–æ—Å–µ—â–µ–Ω–∏–µ –∫—É–ª—å—Ç—É—Ä–Ω—ã—Ö —Å–æ–±—ã—Ç–∏–π, –∞ —Å–æ–≤–µ—Ä—à–∞—é—Ç –ø–æ–∫—É–ø–∫–∏, –≤–æ–∑–º–æ–∂–Ω–æ, —É–∂–µ –±–ª–∏–∂–µ –∫ –∏–Ω—Ç–µ—Ä–µ—Å—É–µ–º–æ–π –¥–∞—Ç–µ.

<div class="alert alert-info"> <b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Å—Ç—É–¥–µ–Ω—Ç–∞:</b> –°–ø–∞—Å–∏–±–æ –∑–∞ –∑–∞–º–µ—á–∞–Ω–∏–µ, –Ω–∏–∂–µ –¥–æ–±–∞–≤–∏–ª –≤—ã–≤–æ–¥—ã –ø–æ—Å–ª–µ –∫–∞–∂–¥–æ–π –∏—Ç–µ—Ä–∞—Ü–∏–∏.</div>

```python
#–†–∞—Å—Å—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–∏–π —á–µ–∫ –ø–æ –≤—Ä–µ–º–µ–Ω–∏ —Å–æ–≤–µ—Ä—à–µ–Ω–∏—è –ø–æ–∫—É–ø–∫–∏
mean_buy = orders_1.pivot_table(index = 'order_month', values = 'revenue', aggfunc = ('count', 'sum')).reset_index()
mean_buy.columns = ['order_month', 'count_buy', 'revenue']
mean_buy['mean_buy'] = mean_buy['revenue'] / mean_buy['count_buy']
mean_buy['order_month'] = mean_buy['order_month'].astype('str')
display(mean_buy.head(10))

#–ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫
plt.figure(figsize = (25, 12))
sns.barplot(data = mean_buy, x = 'order_month', y = 'mean_buy')
plt.xlabel("–ú–µ—Å—è—Ü", fontsize = 25)
plt.ylabel("–°—Ä–µ–¥–Ω–∏–π —á–µ–∫", fontsize = 25)
plt.title("–†–∞–∑–º–µ—Ä —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –ø–æ –º–µ—Å—è—Ü–∞–º", fontsize = 25)
```

**–í—ã–≤–æ–¥**

–°—Ä–µ–¥–Ω–∏–π —á–µ–∫ —Ä–∞—Å—Ç–µ—Ç –∫ –∞–∫—Ç–∏–≤–Ω—ã–º –º–µ—Å—è—Ü–∞–º –Ω–æ—è–±—Ä—é 2017 - —è–Ω–≤–∞—Ä—é 2018 –≥–≥. –°—Ä–µ–¥–Ω–∏–π —á–µ–∫ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç –≤ —ç—Ç–æ—Ç –ø–µ—Ä–∏–æ–¥ –æ—Ç 4 –¥–æ 6 —É.–µ.

<div class="alert alert-danger">
<s><b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞:</b> –í—ã–≤–æ–¥?

<div class="alert alert-info"> <b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Å—Ç—É–¥–µ–Ω—Ç–∞:</b> –î–æ–±–∞–≤–∏–ª</div>

```python
#–†–∞—Å—Å—á–∏—Ç–∞–µ–º, –∫–∞–∫ –º–µ–Ω—è–µ—Ç—Å—è LTV –Ω–∞ –ø–æ–∫—É–ø–∞—Ç–µ–ª—è –ø–æ –∫–æ–≥–æ—Ä—Ç–∞–º
#–î–ª—è —Ä–∞—Å—Å—á–µ—Ç–∞ –∏—Å–ø–æ–ª—å–∑—É–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –ø—Ä–µ–¥—ã–¥—â—É–∏—Ö —à–∞–≥–æ–≤
#–ø–æ—Å—á–∏—Ç–∞–µ–º –¥–ª—è –∫–∞–∂–¥–æ–π –∫–æ–≥–æ—Ä—Ç—ã —Å—É–º–º—É –ø–æ–∫—É–ø–æ–∫
cohort_orders_ltv = orders_1.groupby(['first_order_month', 'order_month']).agg({'revenue': 'sum'}).reset_index()

#–æ–ø—Ä–µ–¥–µ–ª–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–æ–≤—ã—Ö –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –¥–ª—è –∫–∞–∂–¥–æ–π –∫–æ–≥–æ—Ä—Ç—ã
cohort_sizes_first = (orders_1.groupby('first_order_month').agg({'uid': 'nunique'}).reset_index())
cohort_sizes_first.columns = ['first_order_month', 'first_buyers']

#—Å–æ–µ–¥–∏–Ω–∏–º –≤—Å–µ –¥–∞–Ω–Ω—ã–µ –≤ –æ–¥–Ω—É —Ç–∞–±–ª–∏—Ü—É
report_cohort = pd.merge(report_orders_count, cohort_orders_ltv, on = ['first_order_month', 'order_month'])
report_total = pd.merge(report_cohort, cohort_sizes_first, on = 'first_order_month')
#print(report_total.head())

#–ú–∞—Ä–∂–∏–Ω–∞–ª—å–Ω–æ—Å—Ç—å —Å–µ—Ä–≤–∏—Å–∞ –∏–∑ —É—Å–ª–æ–≤–∏—è –∑–∞–¥–∞—á–∏
margin_rate = 1

#—Ä–∞—Å—Å—á–∏—Ç–∞–µ–º –≤–∞–ª–æ–≤—É—é –≤—ã—Ä—É—á–∫—É –∏ –≤–æ–∑—Ä–∞—Å—Ç –∫–æ–≥–æ—Ä—Ç
report_total['gp'] = report_total['revenue'] * margin_rate

report_total['age'] = (report_total['order_month'] - report_total['first_order_month']) / np.timedelta64(1, 'M')
report_total['age'] = report_total['age'].round().astype('int')

#–Ω–∞–π–¥—ë–º LTV. –†–∞–∑–¥–µ–ª–∏–º –≤–∞–ª–æ–≤—É—é –ø—Ä–∏–±—ã–ª—å –∫–æ–≥–æ—Ä—Ç –∑–∞ –∫–∞–∂–¥—ã–π –º–µ—Å—è—Ü –Ω–∞ –æ–±—â–µ–µ —á–∏—Å–ª–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –∫–∞–∂–¥–æ–π –∫–æ–≥–æ—Ä—Ç–µ
report_total['ltv'] = report_total['gp'] / report_total['first_buyers']
display(report_total.head())

report_ltv = report_total.pivot_table(index='first_order_month', columns='age', values='ltv', aggfunc='mean')
report_ltv.cumsum(axis=1).round(2) 
```

```python
#—Ä–∞—Å—Å—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–∏–π ltv –ø–æ –∫–æ–≥–æ—Ä—Ç–∞–º –∑–∞ 6 –º–µ—Å—è—Ü–µ–≤
#–≤ —Ä–∞—Å—Å—á–µ—Ç –≤–æ–∑—å–º–µ–º –∫–æ–≥–æ—Ä—Ç—ã "–ø—Ä–æ–∂–∏–≤—à—ã–µ" –Ω–µ –º–µ–Ω–µ–µ 6 –º–µ—Å—è—Ü–µ–≤
#—Å–æ–∑–¥–∞–¥–∏–º —Å—Ä–µ–∑, –≤ –∫–æ—Ç–æ—Ä—ã–π –æ—Ç–±–µ—Ä–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–º –∫–æ–≥–æ—Ä—Ç–∞–º –∏ –º–µ—Å—è—Ü–∞–º
report_ltv_6cohorts = report_ltv[report_ltv.index <= '2017-12-01']
report_ltv_6month = report_ltv_6cohorts.loc[:, '0': '5']

display(report_ltv_6month.cumsum(axis=1).round(2))
print('–°—Ä–µ–¥–Ω–∏–π ltv –∑–∞ 6 –º–µ—Å—è—Ü–µ–≤:', report_ltv_6month.cumsum(axis=1).round(2).mean().mean())

#–ø–æ—Å—Ç—Ä–æ–∏–º —Ç–µ–ø–ª–æ–≤—É—é –∫–∞—Ä—Ç—É
sns.set(style='white')
plt.figure(figsize=(20, 15))
plt.title('–ò–∑–º–µ–Ω–µ–Ω–∏–µ LTV –≤–æ –≤—Ä–µ–º–µ–Ω–∏ –∑–∞ 6 –º–µ—Å—è—Ü–µ–≤', fontsize = (30))
sns.heatmap(report_ltv_6month.cumsum(axis=1).round(2), annot=True, linewidths=1, linecolor='gray')
plt.ylabel('–ú–µ—Å—è—Ü –ø–µ—Ä–≤–æ–≥–æ –∑–∞–∫–∞–∑–∞')
plt.xlabel('–ú–µ—Å—è—Ü –ø–æ–∫—É–ø–∫–∏')
```

<b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ v2:</b> –û–∫–µ–π, –ø—É—Å—Ç—å –±—É–¥–µ—Ç —Ç–∞–∫.

```python
#–†–∞—Å—Å—á–∏—Ç–∞tv —Å—Ä–µ–¥–Ω–∏–π CAC –Ω–∞ –æ–¥–Ω–æ–≥–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª—è –¥–ª—è –≤—Å–µ–≥–æ –ø—Ä–æ–µ–∫—Ç–∞
#–ø–æ—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –∏ —Å—É–º–º—ã –ø–æ–∫—É–ø–æ–∫ –∑–∞ –∫–∞–∂–¥—ã–π –º–µ—Å—è—Ü
orders_month = report_total.groupby('order_month').agg({'n_buyers': 'sum', 'revenue' : 'sum'}).reset_index()

#–æ–±—ä–µ–¥–∏–Ω–∏–º –¥–∞–Ω–Ω—ã–µ –ø–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª—è–º, –ø–æ–∫—É–ø–∫–∞–º –∏ —Ä–∞—Å—Ö–æ–¥–∞–º
costs_month['month'] = pd.to_datetime(costs_month['month'])
costs_month.columns = ['order_month', 'costs']
orders_costs = pd.merge(orders_month, costs_month, on = 'order_month')

#—Ä–∞—Å—Å—á–∏—Ç–∞–µ–º –°–ê–°
orders_costs['cac'] = (orders_costs['costs'] / orders_costs['n_buyers']).round(2)

display(orders_costs.head(10))
print('–°—Ä–µ–¥–Ω–∏–π CAC –Ω–∞ –æ–¥–Ω–æ–≥–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª—è –∑–∞ –≤—Å–µ –≤—Ä–µ–º—è –ø—Ä–æ–µ–∫—Ç–∞', orders_costs['cac'].mean())

#–æ—Ç–æ–±—Ä–∞–∑–∏–º –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ
orders_costs['order_month'] = orders_costs['order_month'].astype('str')
plt.figure(figsize = (25, 12))
sns.barplot(data = orders_costs, x = 'order_month', y = 'cac', color = 'grey')
plt.xlabel("–ú–µ—Å—è—Ü", fontsize = 25)
plt.ylabel("–†–∞–∑–º–µ—Ä CAC –Ω–∞ –æ–¥–Ω–æ–≥–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª—è", fontsize = 25)
plt.title("–°—Ä–µ–¥–Ω–∏ —Ä–∞–∑–º–µ—Ä –°–ê–° –Ω–∞ –æ–¥–Ω–æ–≥–æ –ø–æ–∫—É–ø–∞—Ç–µ–ª—è –ø–æ –º–µ—Å—è—Ü–∞–º", fontsize = 25)
```

```python
#–†–∞—Å—Å—á–∏—Ç–∞–µ–º CAC –ø–æ —Ä–µ–∫–ª–∞–º–Ω—ã–º –∏—Å—Ç–æ—á–Ω–∏–∫–∞–º
#–æ—Ç–±–µ—Ä–µ–º –ø–µ—Ä–≤—ã–π –∏—Å—Ç–æ—á–Ω–∏–∫ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∫–ª–∏–µ–Ω—Ç–∞
visits_first_source = visits.sort_values('start ts').groupby(['uid'])['source id'].first()

#–æ–±—ä–µ–¥–∏–Ω–∏–º –¥–∞–Ω–Ω—ã–µ
visits_orders = orders_1.merge(visits_first_source, on = 'uid')
display(visits_orders.head())

orders_source = visits_orders.groupby(['first_order_month', 'source id']).agg({'uid' : 'nunique'}).reset_index()
display(orders_source.head())
```

```python
#–†–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ ROMI –ø–æ –∫–æ–≥–æ—Ä—Ç–∞–º –≤ —Ä–∞–∑—Ä–µ–∑–µ –∏—Å—Ç–æ—á–Ω–∏–∫–æ–≤

#–ü–æ–¥–≥–æ—Ç–æ–≤–∏–º –¥–∞–Ω–Ω—ã–µ, —Ä–∞—Å—á–∏—Ç–∞–µ–º –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ –ø–æ–∫–∞–∑–∞—Ç–µ–¥–∏ –ø–æ –∫–æ–≥–æ—Ä—Ç–∞–º –∏ –∏—Å—Ç–æ—á–Ω–∏–∫–∞–º
cohort_end = visits_orders.groupby(['first_order_month', 'source id', 'order_month']).agg({'revenue' : 'sum', 'uid' : 'nunique'}).reset_index()

#–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–µ—Ä–≤—ã—Ö –ø–æ–∫—É–ø–∞—Ç–µ–ª–µ–π –ø–æ –∫–æ–≥–æ—Ä—Ç–∞–º –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∏—Å—Ç–æ—á–Ω–∏–∫–∞
first_buyers = visits_orders.groupby(['first_order_month', 'source id']).agg({'uid' : 'nunique'}).reset_index()

#–æ–±—ä–µ–¥–∏–Ω–∏–º —Ç–∞–±–ª–∏—Ü—ã
cohort_buyers = cohort_end.merge(first_buyers, on = ['first_order_month', 'source id'])
display(cohort_end.head())
display(first_buyers.head())
display(cohort_buyers.head())

costs_month_source.columns = ['first_order_month', 'source id', 'costs']
report_end = cohort_buyers.merge(costs_month_source, on = ['first_order_month', 'source id'])
report_end.columns = ['first_order_month', 'source id', 'order_month', 'revenue', 'n_buyers', 'first_buyers', 'costs']
display(report_end.head(15))
```

```python
# –î–∞–≤–∞–π —Ç–µ–ø–µ—Ä—å —Ä–∞–∑–±–µ—Ä–µ–º—Å—è —Å LTV. –õ–æ–≥–∏–∫–∞ —Ä–∞—Å—á–µ—Ç–∞ —Ç–∞–º —Ç–∞–∫–∞—è –∂–µ –∫–∞–∫ –∏ –≤—ã—à–µ, –≥–¥–µ —Ç—ã –∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞–ª –µ–≥–æ –ø–æ –∫–æ–≥–æ—Ä—Ç–∞–º. 
# –¢–æ–ª—å–∫–æ —Ç–µ–ø–µ—Ä—å —É –Ω–∞—Å –¥–æ–±–∞–≤–∏–ª–∏—Å—å –∏—Å—Ç–æ—á–Ω–∏–∫–∏. –ò—Ö —Ç—ã —Ç–æ–∂–µ –Ω–∞—à–µ–ª –ø—Ä–∞–≤–∏–ª—å–Ω–æ
visits_first_source
```

```python
# –ó–Ω–∞—á–∏—Ç –æ—Å—Ç–∞–ª–æ—Å—å –ø–µ—Ä–µ—Å–æ–±—Ä–∞—Ç—å LTV, –Ω–æ —Å —É—á–µ—Ç–æ–º –∏—Å—Ç–æ—á–Ω–∏–∫–æ–≤. 
```

```python
rev_orders = orders_1.copy() # –°–æ–∑–¥–∞–º –∫–æ–ø–∏—é —Ç–∞–±–ª–∏—Ü—ã, —á—Ç–æ–±—ã –∏–∑–æ–ª–∏—Ä–æ–≤–∞—Ç—å –º–æ—é —á–∞—Å—Ç—å —Ä–∞—Å—á–µ—Ç–∞ –æ—Ç —Ç–≤–æ–µ–≥–æ

# –ù–∞–π–¥–µ–º –ø–µ—Ä–≤—ã–π –º–µ—Å—è—Ü –ø–æ–∫—É–ø–∫–∏ –≤ —Ä–∞–∑—Ä–µ–∑–µ –∏—Å—Ç–æ—á–Ω–∏–∫–æ–≤
rev_orders = rev_orders.join(visits_first_source, on='uid')
first_order_source = rev_orders.groupby(['uid', 'source id']).agg({'buy ts':'min'}).\
                     rename(columns={'buy ts':'first_order_date'}).reset_index()
first_order_source['first_order_month'] = first_order_source['first_order_date'].astype('datetime64[M]')
first_order_source.head()
```

```python
# –ù–∞–π–¥–µ–º —Ä–∞–∑–º–µ—Ä –∫–æ–≥–æ—Ä—Ç —Å —É—á–µ—Ç–æ–º –∏—Å—Ç–æ—á–Ω–∏–∫–æ–≤
cohort_sizes_source = first_order_source.groupby(['first_order_month', 'source id']).agg({'uid':'nunique'}).\
                      rename(columns={'uid':'n_buyers'}).reset_index()
cohort_sizes_source.head()
```

```python
# –ù–∞–π–¥–µ–º –¥–æ—Ö–æ–¥ –ø–æ –∫–æ–≥–æ—Ä–∞—Ç–∞–º —Å —É—á–µ—Ç–æ–º –∏—Å—Ç–æ—á–Ω–∏–∫–æ–≤
cohorts_source = rev_orders.groupby(['first_order_month', 'order_month', 'source id']).agg({'revenue':'sum'}).\
                 reset_index()
cohorts_source.head()
```

```python
# –û–±–µ–¥–∏–Ω—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –∏ –Ω–∞—Ö–æ–¥–∏–º LTV
report_source = pd.merge(cohort_sizes_source, cohorts_source, on=['first_order_month', 'source id'], how='right')

margin_rate = 1
report_source['gp'] = report_source['revenue'] * margin_rate
report_source['age'] = (report_source['order_month'] - report_source['first_order_month']) / np.timedelta64(1, 'M')
report_source['age'] = report_source['age'].round().astype('int')
report_source['ltv_source'] = report_source['gp'] / report_source['n_buyers']
report_source.head()
```

```python
# –ü—Ä–∏–º–µ—Ä –∫–æ–≥–æ—Ä—Ç–Ω–æ–≥–æ LTV –¥–ª—è –ø–µ—Ä–≤–æ–≥–æ –∏—Å—Ç–æ—á–Ω–∏–∫–∞, –±–µ–∑ —É—á–µ—Ç–∞ –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ–π —Å—É–º–º—ã
ltv_source_1 = report_source[report_source['source id'] == 1].pivot_table(index='first_order_month',
                                                           columns='age',
                                                           aggfunc='mean',
                                                           values='ltv_source')
ltv_source_1.cumsum(axis=1)
# –¢–∞–∫, –ø–æ–ª—É—á–∏–ª–∏ —Ç–µ –∂–µ –∑–Ω–∞—á–µ–Ω–∏—è —á—Ç–æ –∏ —É —Ç–µ–±—è. –ó–Ω–∞—á–∏—Ç ROMI –ø–∞–¥–∞–ª –ø–æ –∏–Ω–æ–π –ø—Ä–∏—á–∏–Ω–µ.–ü–æ–µ—Ö–∞–ª–∏ –¥–∞–ª—å—à–µ
```

```python
# –î–æ–±–∞–≤–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏—è CAC –∏ –Ω–∞—Ö–æ–¥–∏–º ROMI
report_source = report_source.merge(source_costs[['source id', 'cac']], on='source id')
```

```python
report_source['romi'] = report_source['ltv_source'] / report_source['cac']
report_source.head()
```

```python
roi_source_1 = report_source[report_source['source id'] == 1].pivot_table(index='first_order_month',
                                                           columns='age',
                                                           aggfunc='mean',
                                                           values='romi')
roi_source_1.cumsum(axis=1)
```
