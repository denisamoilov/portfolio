# Проектная работа. Анализ бизнес-показателей

**Описание задачи**

Необходимо помочь маркетологам снизить расходы — отказаться от невыгодных источников трафика и перераспределить бюджет.
Есть данные Яндекс.Афиши с июня 2017 по конец мая 2018 года:
* лог сервера с данными о посещениях сайта Яндекс.Афиши,
* выгрузка всех заказов за этот период,
* статистика рекламных расходов.

Необходимо изучить:
* как клиенты пользуются сервисом,
* когда делают первые покупки на сайте,
* сколько денег приносит компании каждый клиент,
* когда расходы на привлечение клиента окупаются.

Выручка измеряется в условных единицах — у.е.

Результатом исследования будет:
1. Определение источников трафика, на которые маркетологам стоит делать упор.
2. Выводы, которые будут сформулированы после подсчёта метрик каждого вида: маркетинговых, продуктовых и метрик электронной коммерции.
3. Итоги когортного анализа, определение самых перспективных для компании когорты клиентов.

```python
#определим, сколько в среднем раз в день пользователи заходят на сайт
#посчитаем количество уникальных пользователей в день
users_for_day_uniq = visits.groupby('date').agg({'uid': 'nunique'}).reset_index()

#посчитаем общее количество пользователей в день
users_for_day_count = visits.groupby('date').agg({'uid': 'count'}).reset_index()

#объединим данные и посчитаем среднее количество визитов день
users_for_day = users_for_day_uniq.merge(users_for_day_count, on = 'date')
users_for_day.columns = ['date', 'uniq_uid', 'count_visits']
users_for_day['mean_visits'] = users_for_day['count_visits'] / users_for_day['uniq_uid']

display(users_for_day.head(15))
print("Среднее количество визитов пользователей на сайт в день", users_for_day['mean_visits'].mean())

#посчитаем среднее количество визитов в день в разрезе месяцев, так как они неравновесны
users_for_day['visit_month'] = users_for_day['date'].astype('datetime64[M]')
users_for_day_month = users_for_day.groupby('visit_month').agg({'uniq_uid': 'sum', 'count_visits': 'sum'}).reset_index()
users_for_day_month['mean_visits'] = users_for_day_month['count_visits'] / users_for_day_month['uniq_uid']
display(users_for_day_month.head(15))

#построим график
users_for_day_month['visit_month'] = users_for_day_month['visit_month'].astype('str')
sns.set()
plt.figure(figsize = (10,15))
ax = sns.barplot(data = users_for_day_month, x = 'mean_visits', y = 'visit_month', color = 'salmon', orient = "h")
plt.xlabel("Среднее количество визитов на сайт", fontsize = 15)
plt.ylabel("Дата", fontsize = 15)
plt.title("Среднее количество визитов пользователей на сайт в день", fontsize = 15)

for i in ax.patches:
    ax.text(i.get_width()+.1, i.get_y()+.31, \
            str(round((i.get_width()), 2)), fontsize=15, color='dimgrey')
ax.set_xticks([0, 2])
```

**Вывод**

Среднее количество визитов пользователей на сайт в день составляет 1 раз. Предполагаю, если строить сквозной график по всем дням, он не будет показательным, так как происходит выравнивание между активными осенне-зимними месяцами и летними, поэтому необходимо посмотреть на среднее количество посещений на 1 пользователя по месяцам. Однако, как показывает график, отражающий изменения метрики во времени, сгруппированный по месяцам, картина не изменилась и среднее количество посещений на 1 человека также составляет 1, т.е. количество посещений в активный период было обеспечено за счет уникальный пользователей.

```python
#выделим месяц покупки и месяц первой покупки
visits_1['first_visit_month'] = visits_1['first_visit_date'].astype('datetime64[M]')
visits_1['visit_month'] = visits_1['start ts'].astype('datetime64[M]')

#рассчитаем lifetime
visits_1['cohort_lifetime'] = ((visits_1['visit_month'] - visits_1['first_visit_month']) / np.timedelta64(1, 'M')).round()

#создадим таблицу с когортами
cohort = visits_1.groupby(['first_visit_month', 'cohort_lifetime']).agg({'uid': 'nunique'}).reset_index()
#print(cohort)

initial_users = cohort[cohort['cohort_lifetime'] == 0][['first_visit_month', 'uid']]
#print(initial_users) 
initial_users = initial_users.rename(columns={'uid':'cohort_users'})

cohort = cohort.merge(initial_users, on='first_visit_month') 
display(cohort)

#рассчитаем retention
cohort['retention'] = cohort['uid']/cohort['cohort_users']

#построим сводную таблицу и посчитаем Retention
retention_pivot = cohort.pivot_table(
    index='first_visit_month',
    columns='cohort_lifetime',
    values='retention',
    aggfunc='sum',
)

#построим тепловую карту
sns.set(style='white')
plt.figure(figsize=(20, 15))
plt.title('Cohorts: User Retention', fontsize = (30))
sns.heatmap(retention_pivot, annot=True, fmt='.1%', linewidths=1, linecolor='gray')
plt.show()
```

```python
#Рассчитаем, как меняется LTV на покупателя по когортам
#Для рассчета используем результаты предыдщуих шагов
#посчитаем для каждой когорты сумму покупок
cohort_orders_ltv = orders_1.groupby(['first_order_month', 'order_month']).agg({'revenue': 'sum'}).reset_index()

#определим количество новых покупателей для каждой когорты
cohort_sizes_first = (orders_1.groupby('first_order_month').agg({'uid': 'nunique'}).reset_index())
cohort_sizes_first.columns = ['first_order_month', 'first_buyers']

#соединим все данные в одну таблицу
report_cohort = pd.merge(report_orders_count, cohort_orders_ltv, on = ['first_order_month', 'order_month'])
report_total = pd.merge(report_cohort, cohort_sizes_first, on = 'first_order_month')
#print(report_total.head())

#Маржинальность сервиса из условия задачи
margin_rate = 1

#рассчитаем валовую выручку и возраст когорт
report_total['gp'] = report_total['revenue'] * margin_rate

report_total['age'] = (report_total['order_month'] - report_total['first_order_month']) / np.timedelta64(1, 'M')
report_total['age'] = report_total['age'].round().astype('int')

#найдём LTV. Разделим валовую прибыль когорт за каждый месяц на общее число пользователей в каждой когорте
report_total['ltv'] = report_total['gp'] / report_total['first_buyers']
display(report_total.head())

report_ltv = report_total.pivot_table(index='first_order_month', columns='age', values='ltv', aggfunc='mean')
report_ltv.cumsum(axis=1).round(2) 
```

```python
#рассчитаем ltv, cac, romi
report_end_1['ltv'] = report_end_1['gp'] / report_end_1['first_buyers']
report_end_1['romi'] = report_end_1['ltv'] / report_end_1['cac']
    
    #построим графики    
source_list = report_end_1['source id'].unique()
for source in source_list:

    #ltv по источникам
    report_ltv = report_end_1[report_end_1['source id'] == source].pivot_table(index='first_order_month', columns='age', values='ltv', aggfunc='mean')
    plt.figure(figsize=(20, 15))
    plt.title('Изменение LTV во времени для источника: %i' %source, fontsize = (30))
    sns.heatmap( report_ltv.cumsum(axis=1).round(2), annot=True, linewidths=1, linecolor='gray')
    plt.ylabel('Месяц первого заказа')
    plt.xlabel('Возраст когорты')
    
    #romi по источникам
    report_romi = report_end_1[report_end_1['source id'] == source].pivot_table(index='first_order_month', columns='age', values='romi', aggfunc='mean')
    
    plt.figure(figsize = (25, 15))
    sns.heatmap(data = report_romi.cumsum(axis=1), annot=True, linewidths=1, linecolor='gray')
    plt.xlabel("Возраст когорты", fontsize = 25)
    plt.ylabel("ROMI", fontsize = 25)
    plt.title('Изменение ROMI во времени по когортам для источника: %i' %source, fontsize = (30))
```
