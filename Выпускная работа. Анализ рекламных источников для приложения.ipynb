````python
#посчитаем метрику удержания (Retention Rate)
#выделим дату первой активности пользователя в приложении

first_date = game_actions.groupby('user_id')['event_date'].min()
first_date.name = 'first_date'
user_activity = game_actions.merge(first_date, on='user_id')
display(user_activity.head())
#print(user_activity.info())

#преобразуем тип столбца с датой первой активности в формат даты
user_activity['first_date'] = pd.to_datetime(user_activity['first_date'])
print(user_activity.info())
````

````python
#рассчитаем lifetime пользователя

user_activity['cohort_lifetime'] = (((user_activity['event_date'] - user_activity['first_date']) 
                                     / np.timedelta64(1,'D')).round().astype(int)
                                   )
#сгруппируем данные по когорте и lifetime
cohorts = user_activity.groupby(['first_date', 'cohort_lifetime']).agg({'user_id':'nunique'}).reset_index()

#посчитаем количество пользователей в каждо когорте и добавим эти данные к таблице cohorts
users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_date','user_id']]
users_count = users_count.rename(columns={'user_id':'cohort_users'})
cohorts_user = cohorts.merge(users_count, on='first_date')

display(cohorts_user.head(10))
````

````python
#рассчитаем retention rate, проведем когортный анализ

cohorts_user['retention'] = cohorts_user['user_id'] / cohorts_user['cohort_users']
retention = cohorts_user.pivot_table(index='first_date',columns='cohort_lifetime',values='retention',aggfunc='sum')

#построим тепловую карту
sns.set(style='white')
plt.figure(figsize=(30, 23))
plt.title('Cohorts: User Retention', fontsize = (30))
sns.heatmap(retention, annot=True, fmt='.1%', linewidths=1, linecolor='gray')
plt.show()
````
